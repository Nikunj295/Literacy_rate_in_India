```python
import pandas as pd
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
df_r = pd.read_csv("rural.csv")
df_r['Type'] ="rural"
```

```python
df_u = pd.read_csv("urban.csv")
df_u['Type'] ="urban"
```

```python
# ans["Hindu"] = (df_r['Hindus Male']+df_u['Hindus Male'])/2
States = df_u.iloc[:,1]
Hindu= (df_u.iloc[:, 2]+df_u.iloc[:, 3]+df_r.iloc[:, 3]+df_r.iloc[:, 2])/4
Muslims= (df_u.iloc[:, 5]+df_u.iloc[:, 6]+df_r.iloc[:, 5]+df_r.iloc[:, 6])/4
Chris= (df_u.iloc[:, 8]+df_u.iloc[:, 9]+df_r.iloc[:, 8]+df_r.iloc[:, 9])/4

df = pd.DataFrame([States,Hindu,Muslims,Chris],index=["States","Hindu","Muslims","Chris"])
df = df.T 
df = df.fillna(df.mean())
df = df.drop(35)
```

```python
plt.plot('States', 'Hindu', data=df, linewidth=2)
plt.plot('States', 'Muslims',linestyle='dashed', data=df,  linewidth=2)
plt.plot('States', 'Chris', data=df,linewidth=2)
plt.xticks(rotation=90)
# plt.legend()
plt.legend(loc=4)
ax = plt.axes()        
ax.yaxis.grid()
plt.ylabel("Literacy Rate")
plt.title("Literacy Rate in each Religion in different state of India in 2007")
plt.tight_layout()
```
